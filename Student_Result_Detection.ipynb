```python
from sklearn.model_selection import train_test_split ,GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df.columns
```

```python
df.head(2)
```

```python
features=["study_hours_per_day",'attendance_percentage','mental_health_rating','sleep_hours','part_time_job']
```

```python
target = "exam_score"
```

```python
df_model = df[features+[target]].copy()
```

```python
df_model
```

```python
le = LabelEncoder()
```

```python
df_model["part_time_job"] =le.fit_transform(df_model["part_time_job"])
```

```python
df_model
```

```python
X = df_model[features]
y = df_model[target]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2)
```

```python
len(y_test)
```

```python
len(y_train)
```

```python
from ast import Param
models ={
    "Linear Regression": {"model": LinearRegression(), "params": {}},
    "Ridge": {"model": Ridge(), "params": {}},
    "Decision Tree": {"model": DecisionTreeRegressor(), "params": {"max_depth":[3,5,10],"min_samples_split":[2,5,10]}},
    "RandomForest": {"model": RandomForestRegressor(), "params": {"n_estimators":[50,100],"max_depth":[3,5,10]}}
}
```

```python
best_model = []
```

```python
for name, config in models.items():
    print(f"Training {name}")

    # Create copies to avoid SettingWithCopyWarning and ensure numerical data
    X_train_processed = X_train.copy()
    X_test_processed = X_test.copy()

    # Ensure 'part_time_job' is numerically encoded for training
    if 'part_time_job' in X_train_processed.columns and X_train_processed['part_time_job'].dtype == 'object':
        X_train_processed['part_time_job'] = le.transform(X_train_processed['part_time_job'])
        X_test_processed['part_time_job'] = le.transform(X_test_processed['part_time_job'])

    grid = GridSearchCV(config["model"], config["params"], cv=5, scoring="neg_mean_squared_error")
    grid.fit(X_train_processed, y_train)
    y_pred = grid.predict(X_test_processed)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    best_model.append({
        "model": name,
        "best_params": grid.best_params_,
        "mse": mse,
        "r2": r2
    })
```

```python
result_df = pd.DataFrame(best_model)
```

```python
result_df.sort_values(by="mse")
```

```python
import joblib
best_row = result_df.sort_values(by="mse").iloc[0]
```

```python
best_model__name = best_row["model"]
```

```python
best_model_config = models[best_model__name]
```

```python
final_model = best_model_config["model"]
```

```python
%%writefile app.py
import streamlit as st
import numpy as np
import joblib
from sklearn.preprocessing import LabelEncoder

# Load the trained model
model = joblib.load("final_model.pkl")

# Initialize LabelEncoder for 'part_time_job'
# Fit it with the possible values 'No' and 'Yes' to ensure consistency
# 'No' maps to 0, 'Yes' maps to 1 based on the original LabelEncoder fitting
le = LabelEncoder()
le.fit(["No", "Yes"])

st.title("Student Exam Score Predictor")
st.write("Enter the student's characteristics to predict their exam score.")

# Input widgets
study_hours = st.slider("Study Hours Per Day:", 0.0, 12.0, 2.0)
attendance = st.slider("Attendance Percentage:", 0.0, 100.0, 80.0)
mental_health = st.slider("Mental Health Rating (1-10):", 1.0, 10.0, 5.0)
sleep_hours = st.slider("Sleep Hours Per Day:", 0.0, 12.0, 7.0)
part_time_job_option = st.selectbox("Part Time Job:", ("No", "Yes"))

# Encode 'part_time_job'
ptj_encoded = le.transform([part_time_job_option])[0]

if st.button("Predict Exam Score"):
    input_data = np.array([[study_hours, attendance, mental_health, sleep_hours, ptj_encoded]])
    prediction = model.predict(input_data)[0]
    # Ensure prediction is within a reasonable range (0-100 for exam score)
    prediction = max(0, min(100, prediction))
    st.success(f"Predicted Exam Score: {prediction:.2f}")
```
